```python
from datasets import load_dataset
from tqdm import tqdm
import pandas as pd
from pathlib import Path
import random
```

```python
DATA_DIR = Path("/content/drive/MyDrive/SLM/Data")
MODEL_DIR = Path("/content/drive/MyDrive/SLM/Model")
```

```python
tinystories = load_dataset(
    "roneneldan/TinyStories",
    split="train"
)
```

```python
tinystories = tinystories.select(range(200_000))
```

```python
persona_chat = load_dataset(
    "awsaf49/persona-chat",
    split = "train"
)
```

```python
def format_tinystory(content):

  txt = content["text"] if "text" in content else content["story"]
  return f"<user> Tell me a short story.\n<assistant> {txt.strip()}\n<eos>\n"
```

```python
tinystories_text = [format_tinystory(cnt) for cnt in tinystories]
```

```python
tinystories_text[0]
```

```python
import re
```

```python
def format_personachat(content):
  txt = content["text"].strip()
  if "your persona:" in txt or "partner's persona:" in txt:
    return ""

  txt = re.sub(r"^\d+\s+", "", txt)


  parts = txt.split("\t")

  utterances = []
  for p in parts:
    subparts = p.split("|")
    for sp in subparts:
      s = sp.strip()
      if s:
        utterances.append(s)
  out = []
  is_user = True
  for utt in utterances:
    role = "<user>" if is_user else "<assistant>"
    out.append(f"{role} {utt}")
    is_user = not is_user

  if len(out) < 2:
    return ""

  return "\n".join(out) + "\n<eos>\n"
```

```python
personachat_text = [format_personachat(chat) for chat in persona_chat]
```

```python
personachat_text = [txt for txt in personachat_text if txt.strip()]
```

```python
print("Total formatted PersonaChat samples:", len(personachat_text))
print("\n--- SAMPLE ---\n")
print(personachat_text[0])
```

```python
all_lines = tinystories_text + personachat_text
```

```python
random.shuffle(all_lines)
```

```python
val_frac = 0.01
val_size = int(len(all_lines)* val_frac)
val_size, len(all_lines)
```

```python
val_lines = all_lines[:val_size]
```

```python
train_lines = all_lines[val_size:]
```

```python
(train_path := DATA_DIR / "train.txt").write_text("".join(train_lines), encoding="utf-8")
(val_path := DATA_DIR / "val.txt").write_text("".join(val_lines), encoding="utf-8")
```

```python
print("Train lines:", len(train_lines), "Val lines:", len(val_lines))
print("Saved to", train_path, "and", val_path)
```
